```python
# analysis of single variable:

print("CATEGORICAL VARIABLE")

categorical_colns = ("Customer Segment","Category","Region","Payment Mode","Delivery Status","Auto Reorder") 

for col in categorical_colns:
    print(f"---{col}---")
    print(df[col].value_counts())
    #Visualization

    plt.figure(figsize = (3,3))
    df[col].value_counts().sort_values(ascending = False).plot(kind = "bar",color = "orange")
    plt.title(f"distrubution of {col}")
    plt.xticks(rotation = 30)
    plt.tight_layout()
    plt.show()
```

# Numerical Data analysis

```python
print("\n NUMERICAL ANALYSIS:")

data = ["Quantity","Unit Price","Discount (%)","Sales Amount","Cost Price","Profit","Stock Left","Reorder Quantity"]

for col in data:
    print(f"---{col}---")
    print(f"mean : {df[col].mean():.2f}")
    print(f"median : {df[col].median():.2f}")
    print(f"std dev : {df[col].std():.2f}")
    print(f"min : {df[col].min():.2f}")
    print(f"max : {df[col].max():.2f}")

    # VISUALISATION

    plt.figure(figsize = (8,4))

    plt.subplot(1, 2, 1)
    df[col].hist(bins = 20, color = "blue" , alpha = 0.7)
    plt.title(f"distribution of {col}")
    
    plt.subplot(1, 2, 2)
    df.boxplot(column = col)
    plt.title(f"boxplot of {col}")
    
    plt.tight_layout()
    plt.show()
```

# BIVARIATE ANALYSIS

```python
# Bivariate analysis is statistical analysis defines the relation between two variables 

print("\n...SALES AND PROFIT BY CATEGORY...\n")

category_performance = df.groupby(["Category"]).agg({
    "Sales Amount":"sum",
    "Profit":"sum",
    "Order ID":"count"}).rename(columns = {"Order ID":"Order count"})
category_performance["Profit Margin"] = category_performance["Profit"] / category_performance["Sales Amount"] * 100
print(category_performance,"\n\n")

#Visualization
fig,axes = plt.subplots(2,2,figsize = (15,12))

# sales by category

category_performance["Sales Amount"].plot(kind = "bar", ax=axes[0,0], color="orange", alpha=0.7)
axes[0,0].set_title("Total sales by category")
axes[0,0].tick_params(axis="x",rotation=30)

# profit by category

category_performance["Profit"].plot(kind="bar",ax=axes[0,1], color="red",alpha=0.7)
axes[0,1].set_title("Profit by category")
axes[0,1].tick_params(axis="x", rotation = 30)

#profit margin

category_performance["Profit Margin"].plot(kind = "bar", ax=axes[1,0], color="pink", alpha=0.7)
axes[1,0].set_title("profit margin by category")
axes[1,0].tick_params(axis="x", rotation=30)

#order count

category_performance["Order count"].plot(kind="bar", ax=axes[1,1], color="black",alpha=0.7)
axes[1,1].set_title("order count")
axes[1,1].tick_params(axis="x",rotation=30)

plt.tight_layout()
plt.show()

```

# Region Performance

```python
print("REGIONAL PERFORMANCE \n")

region_performance = df.groupby(['Region']).agg({
    'Sales Amount' : 'sum',
    'Profit' : 'sum',
    'Order ID' : 'count'
}).rename(columns={'Order ID':'Order count'})
region_performance["Profit Margin"] = region_performance['Profit'] / region_performance['Sales Amount']*100
print(region_performance,"\n\n")

# Visualization

plt.figure(figsize = (12,8))

# visualize region by sales
plt.subplot(2,2,1)
region_performance["Sales Amount"].plot(kind = "pie", autopct="%1.1f%%")
plt.title("Sales by region")

#visuals by region by sakles

plt.subplot(2,2,2)
region_performance["Profit"].plot(kind="pie", autopct="%1.1f%%")
plt.title("profit by region")

plt.subplot(2,2,3)
sns.barplot(x= region_performance.index,y="Order count", data=region_performance)
plt.title("order count by region")
plt.xticks(rotation=30)

plt.subplot(2,2,4)
sns.barplot(x=region_performance.index,y="Profit Margin", data=region_performance)
plt.title("order count by profit margin (%)")
plt.xticks(rotation=30)

plt.tight_layout()
plt.show()
```

# Time Series Analysis 

```python
print("TIME SERIES ANALYSIS")

# Extract time components
df["Order Month"] = df["Order Date"].dt.to_period("M")
df["Order Year"] = df["Order Date"].dt.to_period("Y")

monthly_trend = df.groupby(["Order Month"]).agg({
    "Sales Amount":"sum",
    "Profit":"sum",
    "Order ID":"count"
}).rename(columns = {"Order ID" : "Order count"})

print("monthly_trend")
print(monthly_trend.tail(10),"\n\n")

# Visualization

plt.figure(figsize=(15,10))

plt.subplot(3,1,1)
monthly_trend["Sales Amount"].plot(kind="line" , color="green",marker="o")
plt.title("Monthly sales analysis")
plt.ylabel("sales amount")
plt.grid(True)

plt.subplot(3,1,2)
monthly_trend["Profit"].plot(kind="line",color="darkblue",marker="o")
plt.title("Monthly profit")
plt.ylabel("Profit")
plt.grid(True)

plt.subplot(3,1,3)
monthly_trend["Order count"].plot(kind="line",color="grey",marker="o")
plt.title("monthly order count")
plt.ylabel("Order count")
plt.grid(True)

plt.tight_layout()
plt.show()
```


# Advance Analysis

```python
df.head(1)
```

```python
# Region-wise Delivery Status
region_status = df.groupby("Region")["Delivery Status"].value_counts().unstack(fill_value=0)

#  Category-wise Status
category_status = df.groupby("Category")["Delivery Status"].value_counts().unstack(fill_value=0)

#  Customer Segment Status
segment_status = df.groupby("Customer Segment")["Delivery Status"].value_counts().unstack(fill_value=0)

#  Supplier Cancel & Return Rate (Top 10)
supplier_status = df.groupby("Supplier Name")["Delivery Status"].value_counts(normalize=True).unstack(fill_value=0)
supplier_top = supplier_status.sort_values(by=["Cancelled","Returned"], ascending=False).head(10)

# Display tables
print("\n📍 Region-wise Delivery Status\n", region_status)
print("\n📦 Category-wise Delivery Status\n", category_status)
print("\n👤 Customer Segment Delivery Status\n", segment_status)
print("\n🚚 Top 10 Suppliers With Cancellation/Return Issues\n", supplier_top)

# =======================
#      PLOTS
# =======================

# 1️ Region-wise
region_status.plot(kind="bar", figsize=(6,4))
plt.title("Region-wise Delivery Status")
plt.xlabel("Region")
plt.ylabel("Order Count")
plt.show()

# 2️ Category-wise
category_status.plot(kind="bar", figsize=(6,4))
plt.title("Category-wise Delivery Status")
plt.xlabel("Category")
plt.ylabel("Order Count")
plt.show()

# 3️ Segment-wise
segment_status.plot(kind="bar", figsize=(6,4))
plt.title("Customer Segment Delivery Status")
plt.xlabel("Customer Segment")
plt.ylabel("Order Count")
plt.show()

# 4️ Suppliers With Highest Cancellation/Return Rate
supplier_top[["Cancelled","Returned"]].plot(kind="bar", figsize=(5,5))
plt.title("Suppliers With Highest Cancellations & Returns")
plt.xlabel("Supplier Name")
plt.ylabel("Proportion")
plt.show()
```

```python
print("CORRELATION ANALYSIS")

# select numerical column for corerelation

numerical_df =df[["Quantity","Unit Price","Discount (%)","Sales Amount","Cost Price","Profit"]]

# correlation matrix

correlation_matrix = numerical_df.corr() 
print("correlation_matrix:")
print(correlation_matrix,"/n/n")

# heat map visualize

plt.figure(figsize=(13,13))
sns.heatmap(correlation_matrix,annot=True,cmap="coolwarm",center=0,square=True,linewidth=0.5)
plt.title("correlation heatmap" , fontsize=16,fontweight="bold")
plt.tight_layout()
plt.show()
```

# Customer Segment Analysis

```python
print("CUSTOMER SEGMENT ANALYSIS")

cust_seg = df.groupby(df["Customer Segment"]).agg({
    "Sales Amount":["sum","mean"],
    "Profit":["sum","mean"],
    "Order ID":"count",
    "Discount (%)":"mean"
}).round(2)

cust_seg.columns = ["total sales","avg sales","total profit","avg profit","order count","avg discount"]
print(cust_seg,"/n/n")

#Visualization

fig,axes=plt.subplots(2,2,figsize=(15,10))

cust_seg["total sales"].plot(kind="bar",ax=axes[0,0],color="yellow")
axes[0,0].set_title("TOTAL SALES BY CUST SEG")
axes[0,0].tick_params(axis="x",rotation=30)

cust_seg["total profit"].plot(kind="bar",ax=axes[0,1],color="blue")
axes[0,1].set_title("TOTAL PROFIT BY CUST SEG")
axes[0,1].tick_params(axis="x",rotation=30)

cust_seg["avg profit"].plot(kind="bar",ax=axes[1,0],color="red")
axes[1,0].set_title("AVG PROFIT BY CUST SEG")
axes[1,0].tick_params(axis="x",rotation=30)

cust_seg["avg discount"].plot(kind="bar",ax=axes[1,1],color="pink")
axes[1,1].set_title("AVG DISCOUNT")
axes[1,1].tick_params(axis="x",rotation=30)

plt.tight_layout()
plt.show()
    
```

# Key insides and Summary

```python
# Overall matrix
total_sales=df["Sales Amount"].sum()
total_profit=df["Profit"].sum()
total_order=df["Order ID"].nunique()
total_profit_margin=(total_profit/total_sales)*100

print(f"Overall Performance")
print(f"total_sales:{total_sales:.2f}")
print(f"total_profit:{total_profit:.2f}")
print(f"total_order:{total_order}")
print(f"total_profit_margin:{total_profit_margin:.2f}")

#Visualization
values = [total_sales, total_profit]
labels = ["Total Sales", "Total Profit"]

plt.bar(labels, values)
plt.title("Total Sales vs Total Profit")
plt.ylabel("Amount in cr")
plt.show()

```

```python
#Top performing categories

top_categories = df.groupby("Category")["Profit"].sum().sort_values(ascending=False)
print(f"top performing categories")
for i,(category,profit) in enumerate(top_categories.items(),1):
    
    print(f"{i}.{category}:{profit:,.2f}")
```

```python
#Regional performance

best_region = df.groupby("Region")["Profit"].sum().idxmax()
print(f"\n best performing region:{best_region}")
```

```python
# Delivery performance 

delivery_stats=df["Delivery Status"].value_counts(normalize = True)*100
print(f"delivery status")
for status,percentage in delivery_stats.items():
    print(f"{status}:{percentage:,.1f}%")
```

```python
# Customer segment

best_segment = df.groupby("Customer Segment")["Profit"].sum().idxmax()
print(f"most profitable customer segment:{best_segment}")
```

# Export data

```python
analysis = {
    "total_sales" : total_sales,
    "total_profit" : total_profit,
    "total orders" : total_order,
    "total_profit_margin" : total_profit_margin,
    "top_category" : top_categories.index[0],
    "best_region" : best_region,
    "best_customer_segment":best_segment
}
df1=pd.DataFrame(["analysis"])
df1.to_csv("EDA_analysis_summary.csv",index=False)
print("EDA_analysis_summary.csv")
```
